Q1: In this question you have to scrape data using the filters available on the webpage You have to use the location and salary filter.

You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results.

You have to scrape the job-title, job-location, company name, experience required.

The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs

The task will be done as shown in the below steps:

1. first get the web page https://www.naukri.com/

2. Enter ‚ÄúData Scientist‚Äù in ‚ÄúSkill, Designations, and Companies‚Äù field.

3. Then click the search button.

4. Then apply the location filter and salary filter by checking the respective boxes

5. Then scrape the data for the first 10 jobs results you get.

6. Finally create a dataframe of the scraped data.

```python
def Naukri_JobSearch(url):
    
    # Initializing WebDriver for Chrome and Opening the specified URL  
    driver=webdriver.Chrome()   
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)

    
    # Initialize empty lists for required data storing
    job_titles=[]
    company_names=[]
    years_of_experience=[]
    job_decscription=[]
    
    
    #To input 'Data Scientist' keyword in the 'Enter skills / designations / companies' field
    designation=wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@class="suggestor-input "]'))).send_keys('Data Scientist')
    
    #Code will find the search icon and press click
    search_button=driver.find_element(By.XPATH,'//div[@class="qsbSubmit"]').click()
    
    
    #This code will select the salary filter as '3-6 Lakhs'
    salaries=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//span[@class="styles_ellipsis__cvWP1 styles_filterLabel__jRP04"]')))
    for salary in salaries:
        if salary.text=='3-6 Lakhs':
            salary.click()
            break
    
    #This code will select the location filter as 'Delhi / NCR'
    locations=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//span[@class="styles_ellipsis__cvWP1 styles_filterLabel__jRP04"]')))
    for Delhi_NCR in locations:
        if Delhi_NCR.text=='Delhi / NCR':
            Delhi_NCR.click()
            break
    
    
    #This code will find all the job posting elements and assign it to a variable
    job_posting_frame=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]')))
    
    #This code will loop through each element of 'job_posting_frame' until index value returns 10
    for index,job_postings in enumerate(job_posting_frame):
        if index==10:
            break
        
        #Grabbing the title of the job and append it in the job_titles list
        title=job_postings.find_element(By.XPATH,'.//a[@class="title "]')
        job_titles.append(title.text)
        
        #Grabbing the company name of the job and append it in the company_names list
        company=job_postings.find_element(By.XPATH,'.//span[@class=" comp-dtls-wrap"]/a[1]')
        company_names.append(company.text)
        
        #Grabbing the experience requires for the job and append it in the years_of_experience list
        experience=job_postings.find_element(By.XPATH,'.//span[@class="expwdth"]')
        years_of_experience.append(experience.text)
        
        #Grabbing the job summary and append it in the job_decscription list
        summary=job_postings.find_element(By.XPATH,'.//span[@class="job-desc ni-job-tuple-icon ni-job-tuple-icon-srp-description"]')
        job_decscription.append(summary.text)
        
    
    #Creating a dataframe containing job information
    df_naukri=pd.DataFrame({'Job Title':job_titles,'Company Name':company_names,'Years of experience':years_of_experience,'Job Description':job_decscription},index=range(1,len(job_titles)+1))
    df_naukri=df_naukri.head(10)
    
    #Return dataframe as output 
    return df_naukri


    

Naukri_JobSearch('https://www.naukri.com/')
```

Q2: Write a python program to scrape data for ‚ÄúData Scientist‚Äù Job position in ‚ÄúBangalore‚Äù location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:

1. First get the webpage https://www.shine.com/

2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.

3. Then click the searchbutton.

4. Then scrape the data for the first 10 jobs results you get.

5. Finally create a dataframe of the scraped data.

```python
def Shine_JobSearch(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)
    
    # Initialize empty lists for required data storing
    job_title=[]
    company_names=[]
    job_location=[]
    required_experience=[]
    
    #Activating the 'Job title, skills' field by clicking it
    wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@class="input"]'))).click()
    
    #Code to input 'Data Analyst' in the 'Job title, skills' field
    wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@class="form-control  "]'))).send_keys('Data Analyst')
    
    #Code to input 'Bangalore' in the 'Location' field
    wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@id="id_loc"]'))).send_keys('Bangalore')
    
    #Code to click the search button
    driver.find_element(By.XPATH,'//button[@class=" btn btn-secondary undefined"]').click()
    
    #This code will find all the job posting elements and assign it to a variable job_posting_frame
    job_posting_frame=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="jobCard_jobCard__jjUmu  white-box-border jobCard"]')))
    
    #This code will loop through each element of 'job_posting_frame' until index value returns 10
    for index,job_postings in enumerate(job_posting_frame):
        
        if index == 10:
            break
        
        #Grabbing the title of the job and append it in the title list
        title=job_postings.find_element(By.XPATH,'.//div[@class="jobCard_jobCard_features__wJid6"]/following-sibling::h2/a')
        job_title.append(title.text)
        
        #Grabbing the company name of the job and append it in the company list
        company=job_postings.find_element(By.XPATH,'.//div[@class="jobCard_jobCard_cName__mYnow"]/span')
        company_names.append(company.text)
        
        #Grabbing the company name of the job and append it in the company_names list
        location=job_postings.find_element(By.XPATH,'.//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
        job_location.append(location.text)
        
        #Extracting the location keyword from the list
        job_location=[location.split('\n')[0] for location in job_location]
    
        #Grabbing the experience requires for the job and append it in the years_of_experience list
        experience=job_postings.find_element(By.XPATH,'.//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
        required_experience.append(experience.text)
        
   
    #Creating a dataframe containing job information
    df_shine=pd.DataFrame({'Job Title':job_title,'Company Name':company_names,'Location':job_location,'Experience':required_experience},index=range(1,len(job_title)+1))
    
    #Return dataframe as output 
    return df_shine
    
    
    
Shine_JobSearch('https://www.shine.com/')
```

    1. Rating

    2. Review summary

    3. Full review

    4. You have to scrape this data for first 100reviews.
Note: All the steps required during scraping should be done through code only and not manually.

```python
def Flipkart_iPhone11_Review(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)
    
    # Initialize empty lists for required data storing
    customer_ratings=[]
    product_summary=[]
    full_review=[]
    
    #Creating a loop which iterates for 10 times to visit 10 pages of iPhone review
    for page in range(0,10):
        
        #This code will find the elements which encompasses all the review elements and assign it to a variable customer_review_frame
        customer_review_frame=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="col _2wzgFH K0kLPL"]')))
        
        
        for customer_review in customer_review_frame:
            
            #Code to extract the user ratings 
            ratings=customer_review.find_element(By.XPATH,'.//div[@class="_3LWZlK _1BLPMq"]')
            customer_ratings.append(ratings.text)
    
            #Code to extract the user summary 
            summary=customer_review.find_element(By.XPATH,'.//p[@class="_2-N8zT"]')
            product_summary.append(summary.text)
        
            #Code to extract the full review
            reviews=customer_review.find_element(By.XPATH,'.//div[@class="t-ZTKy"]/div/div')
            full_review.append(reviews.text)
    
        
        #Code to click the next button
        wait.until(EC.visibility_of_element_located((By.XPATH,'//a[@class="_1LKTO3" and contains(span, "Next")]'))).click()
        
        #Stopping the execution of program for 5 seconds
        time.sleep(5)
    
    #Creating a dataframe containing iPhone review
    df_iphone_review=pd.DataFrame({'Customer ratings':customer_ratings,'Review summaries':product_summary,'Full reviews':full_review},index=range(1,len(customer_ratings)+1))
    
    #Return dataframe as output 
    return df_iphone_review
    
    
    
    
Flipkart_iPhone11_Review('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')
```

Q4: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for ‚Äúsneakers‚Äù in the search field.
You have to scrape 3 attributes of each sneaker:
1. Brand
2. Product Description
3. Price

As shown in the below image, you have to scrape the above attributes.

```python
def Flipkart_Sneakers(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)

    # Initialize empty lists for required data storing
    brand_names=[]
    product_description=[]
    product_price=[]
    
    # Code to input 'Sneakers' in search field
    wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@class="Pke_EE"]'))).send_keys('Sneakers')
    
    #Clicks the search button
    driver.find_element(By.XPATH,'//button[@class="_2iLD__"]').click()
        
    #Creating a loop which iterates for 3 times to visit 3 pages of sneaker's list 
    for page in range(3):
        
        #Captures the elements where all the product detail elements are available and initialized it
        product_details=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="_2B099V"]')))
        
        #Creating a loop which iterates through all the elements assigned to product_details
        for sneakers in product_details:
            
            #Finds and append brand name to list
            brand=sneakers.find_element(By.XPATH,'.//div[@class="_2WkVRV"]')
            brand_names.append(brand.text)
            
            #Finds and append product summary to list. Used try...except for elements that are not available 
            try:
                product_summary=sneakers.find_element(By.XPATH,'.//div[@class="_2WkVRV"]/following-sibling::a')
                product_description.append(product_summary.text)
            except NoSuchElementException:
                product_description.append('Not available')
            
            #Finds and append product price to list. Used try...except for elements that are not available 
            try:
                price=sneakers.find_element(By.XPATH,'.//div[@class="_30jeq3"]')
                product_price.append(price.text)
            except NoSuchElementException:
                product_price.append('Not available')

        #Clicks the next link to goto the  next page
        driver.find_element(By.XPATH,'//a[@class="_1LKTO3" and contains(span,"Next")]').click()
        
        #Stopping the execution of program for 5 seconds
        time.sleep(5)
                                             
    #Creating a dataframe containing sneaker's list
    df_flipkartSneakers=pd.DataFrame({'Brands':brand_names,'Product description':product_description,'Price':product_price},index=range(1,len(brand_names)+1))
    
    #Selecting the first 100 sneakers    
    df_flipkartSneakers=df_flipkartSneakers.head(100)
        
    #Return dataframe as output 
    return df_flipkartSneakers
   
    
Flipkart_Sneakers('https://www.flipkart.com/')
```

After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:

```python
def Amazon_Laptop(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)
    
    # Initialize empty lists for required data storing
    laptop_name=[]
    laptop_rating=[]
    laptop_price=[]
    no_of_reviews=[]
    
    #Code to input 'Laptop' in the search box and pressing 'ENTER' through Keys.ENTER
    searchBox=wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@id="twotabsearchtextbox"]')))
    searchBox.send_keys('Laptop')
    searchBox.send_keys(Keys.ENTER)
    
    
    #Code to find the filter 'Intel Core i7' and selecting the same by clicking on the element
    wait.until(EC.visibility_of_element_located((By.XPATH,'//span[@class="a-size-base a-color-base" and contains(text(),"Intel Core i7")]'))).click()
    
    
    #Capturing the elements where all the required elements are available
    products_details_elements=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="a-section a-spacing-small a-spacing-top-small"]')))
    
    #Removing the 1st element form products_details_elements list
    products_details_frame=products_details_elements[1:]
    
    #Creating a loop which iterates through all the elements assigned to products_details_frame
    for index,products in enumerate(products_details_frame):
        if index==10: #IF condition is used to break from the loop once index value is 10 i.e after looping for 10 times 
            break
        
        #Finds and append product name to list
        name=products.find_element(By.XPATH,'.//span[@class="a-size-medium a-color-base a-text-normal"]')
        laptop_name.append(name.text)
    
        #Finds and append ratings to list. Used try...except for elements that are not available
        try:
            ratings=products.find_element(By.XPATH,'.//span[@class="a-icon-alt"]')
            laptop_rating.append(ratings.get_attribute("innerText"))
        except NoSuchElementException:
            laptop_rating.append('Not available')
            
        #Finds and append no. of users to list. Used try...except for elements that are not available
        try:
            users=products.find_element(By.XPATH,'.//span[@class="a-size-base s-underline-text"]')
            no_of_reviews.append(users.text)
        except NoSuchElementException:
            no_of_reviews.append('Not available')
        
        #Finds and append price to list. Used try...except for elements that are not available
        try:    
            price=products.find_element(By.XPATH,'.//span[@class="a-price-whole"]')
            laptop_price.append(price.text)
        except NoSuchElementException:
            laptop_price.append('Not available')
        
    
    #Creating a dataframe containing sneaker's list
    df_laptop_details=pd.DataFrame({"Laptop description":laptop_name,"Price":laptop_price,"Ratings":laptop_rating,"No of reviews":no_of_reviews},index=range(1,len(laptop_name)+1))
    df_laptop_details.head(10)
    
    #Return dataframe as output 
    return df_laptop_details

    
    
    
Amazon_Laptop('https://www.amazon.in/')
```

Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on Top Quote
3. Than scrap a) Quote b) Author c) Type Of Quotes


```python
def AZQuotes(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)
    
    # Initialize empty lists for required data storing
    author_quotes=[]
    authors=[]
    type_of_quotes=[]
    
    #Locates the element Top Quotes and click on it 
    wait.until(EC.visibility_of_element_located((By.XPATH,'//a[contains(text(),"Top Quotes")]'))).click()
 
    
    #Creating a loop which iterates for 10 times to visit 10 pages of top quote's list 
    for page in range(10):
        
        #Captures the elements where all the quotes elements are available and initialized it 
        quote_details=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="wrap-block"]')))
        
        #Iterates through each element of quote_details
        for quote in quote_details:
            
            #Scraping quotes from authors
            quotes=quote.find_element(By.XPATH,'.//a[@class="title"]')
            author_quotes.append(quotes.text)
        
        
            #Scraping author names
            name=quote.find_element(By.XPATH,'.//div[@class="author"]/a')
            authors.append(name.text)
        
            #Scraping types of quotes
            quote_type=quote.find_element(By.XPATH,'.//div[@class="tags"]')
            type_of_quotes.append(quote_type.text)
        
        #Find the next element and clicks on it
        try:
            driver.find_element(By.XPATH,'//li[@class="next"]').click()
            driver.implicitly_wait(5)
        except NoSuchElementException:
            break
        
    #Creating a dataframe containing quotes's list    
    df_quotes=pd.DataFrame({'QUOTES':author_quotes, 'Authors':authors, 'Type of quote':type_of_quotes},index=range(1,len(author_quotes)+1))
    
    #Return dataframe as output 
    return df_quotes
    
    
AZQuotes('https://www.azquotes.com/')
```

Q7: Write a python program to display list of respected former Prime Ministers of India (i.e. Name, Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1
Scrap the mentioned data and make the DataFrame


```python
def ListOfPrimeMisnister(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)
    
    # Initialize empty lists for required data storing
    pm_names=[]
    born_dead=[]
    terms_of_office=[]
    remark=[]
    
    #Capturing the table of list of PM and assigning it to a variable
    pm_table_list = wait.until(EC.visibility_of_element_located((By.XPATH,'//div[@class="TableData"][1]')))
    
    #Capturing all the row elements leaving the 1st row position
    table_rows = pm_table_list.find_elements(By.XPATH,'.//tr[position()>1]')
    
    #Looping through all the rows of the table
    for rows in table_rows:
        
        #Capturing the column elements from the row
        column=rows.find_elements(By.XPATH,'.//td')
        
        #Appending the column text in the PM name,born-dead,terms of office, remarks list by mentioning the position of columns
        pm_names.append(column[1].text)
        born_dead.append(column[2].text)
        terms_of_office.append(column[3].text)
        remark.append(column[4].text)
    
    #Replacing '\n' with ',' from the list 'terms_of_office'
    terms_of_office=[item.replace('\n',',') for item in terms_of_office]
    
    #Creating a dataframe containing PM details    
    df_PM_List=pd.DataFrame({'PM Name':pm_names,'Born-Dead':born_dead,'Term of office':terms_of_office,'Remark':remark},index=range(1,len(pm_names)+1))
    
    #Return dataframe as output 
    return df_PM_List

    
ListOfPrimeMisnister('https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1')
```

Q8: Write a python program to display list of 50 Most expensive cars in the world (i.e. Car name and Price) from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ‚Äô50 most expensive cars‚Äô
3. Then click on 50 most expensive cars in the world..
4. Then scrap the mentioned data and make the dataframe.

```python
def FiftyMostExpensiveCars(url):
    
    # Initializing WebDriver for Chrome and opening the specified URL
    driver=webdriver.Chrome()
    driver.get(url)
    
    # Initialize WebDriverWait with a timeout of 10 seconds
    wait=WebDriverWait(driver,10)
    
    # Initialize empty lists for required data storing
    car_names_list=[]
    car_prices_list=[]
    
    #Capturing the search button and then search 50 most expensive cars 
    search_button = wait.until(EC.visibility_of_element_located((By.XPATH,'//input[@class="m1-search-panel-input m1-search-form-text"]')))
    search_button.click()
    search_button.send_keys("50 most expensive cars")
    search_button.send_keys(Keys.ENTER)
    
    #Click the link of 50 most expensive cars in the world
    wait.until(EC.visibility_of_element_located((By.XPATH,'//a[contains(text(),"50 Most Expensive Cars In The World")]'))).click()
    
    
    #Locates the elements that contains car names 
    car_names=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//h3[@class="subheader"]')))
    
    #Iterates through the car name elements  
    for index,car in enumerate(car_names):
        if index==50:
            break
        car_names_list.append(car.text)
    
    #Locates the elements that contains car price 
    car_price=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//h3[@class="subheader"]/following-sibling::p[1]')))
    
    #Iterates through the car price elements
    for index,price in enumerate(car_price):
        if index==50:
            break
        car_prices_list.append(price.text)
    
    #Filtering the car_prices_list by spliting it
    car_prices_list=[price.split('$')[1] for price in car_prices_list]
    
    #Creating a dataframe containing 50 most expensive cars     
    df_cars_prices=pd.DataFrame({'Car Names':car_names_list,'Price($)':car_prices_list},index=range(1,len(car_names_list)+1))
    
    #Return dataframe as output 
    return df_cars_prices



FiftyMostExpensiveCars('https://www.motor1.com/')
```
