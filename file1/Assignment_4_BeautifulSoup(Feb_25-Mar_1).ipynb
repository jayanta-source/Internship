Question 1: Write a python program to display IMDB’s Top rated 100 Indian movies’ data https://www.imdb.com/list/ls056092300/ (i.e. name, rating, year of release) and make data frame.

```python
async def Top100movies(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)

    await response.html.arender(sleep=10)

    imdb_soup = BeautifulSoup(response.html.html,'lxml')
    
    film_title=[]
    ratings=[]
    release_year=[]
    
    for name_year_rating in imdb_soup.find_all('div',class_='lister-item-content'):
        film_title.append((name_year_rating.find('a')).text)
        release_year.append((name_year_rating.find('span',class_='lister-item-year text-muted unbold')).text)
        ratings.append((name_year_rating.find('span',class_='ipl-rating-star__rating')).text)
        
    
    df=pd.DataFrame({'Movie Names':film_title,'Year of Release':release_year, 'Ratings':ratings})
    return df 
    
    
await Top100movies('https://www.imdb.com/list/ls056092300/')
```

Question 2: Write a python program to scrape product name, price and discounts from https://peachmode.com/search?q=bags

```python
async def Peachmode_bags(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)

    await response.html.arender(sleep=5)

    Bags_soup = BeautifulSoup(response.html.html,'lxml')
    
    product_name=[]
    price=[]
    discount=[]
    
    for items in Bags_soup.find_all('div', class_='product-bottom'):
        product_name.append((items.find('div',class_='product-title')).text)
        price.append((items.find('span',class_='price st-money money done')).text)
        discount.append((items.find('span',class_='discounted_price st-money money done')).text)
        
    df=pd.DataFrame({'Product Name':product_name,'Price':price,'Discounted Price':discount})
    return df
    
    

await Peachmode_bags('https://peachmode.com/search?q=bags')
```

Question 3: Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

    a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
    b) Top 10 ODI Batsmen along with the records of their team and rating.
    c) Top 10 ODI bowlers along with the records of their team and rating.

```python
# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

async def ODI_Top10Team(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)

    await response.html.arender(sleep=10)

    soup = BeautifulSoup(response.html.html,'lxml')
    
    position=[]
    team_names=[]
    match_played=[]
    points=[]
    rating=[]
    
    for table_body in soup.find_all('div',class_='si-table-body'):
        for table_row in table_body.find_all('div',class_='si-table-row'):
                
            for pos in table_row.find_all('div',class_='si-table-data si-pos'):
                position.append((pos.find('span',class_='si-text')).text)
                
            for names in table_row.find_all('h3',class_='si-team-name'):
                team_names.append((names.find('span',class_='si-fname si-text')).text)
                
            for matched in table_row.find_all('div',class_='si-table-data si-matches'):
                match_played.append((matched.find('span',class_='si-text')).text)
                
            for point in table_row.find_all('div',class_='si-table-data si-pts'):
                points.append((point.find('span',class_='si-text')).text)
                
            for rate in table_row.find_all('div',class_='si-table-data si-rating'):
                rating.append((rate.find('span',class_='si-text')).text)    
           
                        
    df=pd.DataFrame({'Position':position, 'Country':team_names,'Total Match Played':match_played,'Points scored':points,'Ratings':rating})
    df=df.head(10)
    return df



await ODI_Top10Team('https://www.icc-cricket.com/rankings/team-rankings/mens/odi')
```

```python
# Top 10 ODI Batsmen along with the records of their team and rating.

async def ODI_top10Batsman(url):
    session = AsyncHTMLSession()
    response = await session.get(url)

    await response.html.arender(sleep=10)

    soup = BeautifulSoup(response.html.html,'lxml')
    
    position=[]
    fname=[]
    lname=[]
    counrty=[]
    rating=[]
    count=0
    
    for table_body in soup.find_all('div',class_='si-table-body'):
        for table_row in table_body.find_all('div',class_='si-table-row'):
            
            for pos in table_row.find_all('div',class_='si-table-data si-pos'):
                position.append((pos.find('span',class_='si-text')).text)
                
            for names in table_row.find_all('h3',class_='si-player-name'):
                
                try:
                    fname.append((names.find('span', class_='si-text si-fname')).text)
                except AttributeError:
                    fname.append('')
                    
                lname.append((names.find('span', class_='si-text si-lname')).text)
                
            for conuntries in table_row.find_all('h3',class_='si-team-name'):
                counrty.append((conuntries.find('span',class_='si-fname si-text')).text)
                
                
            for rate in table_row.find_all('div',class_='si-table-data si-rating'):
                rating.append((rate.find('span',class_='si-text')).text)
            
                
    player_names=[first+' '+last for first, last in zip(fname,lname)]        
                        
    df=pd.DataFrame({'Position':position, 'Player Name':player_names,'Country':counrty,'Ratings':rating})
    df=df.head(10)
    return df


await ODI_top10Batsman('https://www.icc-cricket.com/rankings/batting/mens/odi')
```

Question 4: Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the heading, date, content and the likes for the video from the link for the youtube video from the post.

```python
async def Patreon(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)

    await response.html.arender(sleep=10)

    patreon_soup = BeautifulSoup(response.html.html,'lxml')
    
    headings=[]
    date=[]
    content=[]
    likes=[]
    
    for posts in patreon_soup.find_all('div',class_='sc-jvvksu fTaUtt'):
        headings.append((posts.find('span', class_='sc-1hmaiaw-0 fIBzIT')).text)
        date.append((posts.find('a',class_='sc-jQrDum bMyBFw')).text)
        content.append((posts.find('div',class_='sc-1ye87qi-0 bCBphS')).text)
        
        try:
            likes.append((posts.find('span', class_='sc-fFeiMQ hGHQwf')).text)
        except AttributeError:
            likes.append("Not available right now!")
        
    
    df=pd.DataFrame({'Title':headings,'Release Date':date,'Summary':content,'User likes':likes})
    return df


await Patreon('https://www.patreon.com/coreyms')
```

Question 5: Write a python program to scrape house details from mentioned URL. It should include house title, location, area, EMI and price from https://www.nobroker.in/. Enter three localities which are Indira Nagar, Jayanagar, Rajaji Nagar.

```python
async def NoBroker(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)

    await response.html.arender(sleep=10)

    soup = BeautifulSoup(response.html.html,'lxml')
    
    house_names=[]
    location=[]
    price=[]
    emi=[]
    
    for house_info in soup.find_all('div',class_='flex flex-col flex-2 w-pe mt-1.8px po:justify-center po:p-1p po:mt-0 w-70pe'):
        
        house_names.append((house_info.find('h2',class_='heading-6 flex items-center font-semi-bold m-0')).text)
        
        location.append((house_info.find('div',class_='mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95')).text)
    
    
    for house_price in soup.find_all('div',class_='p-1.5p flex border-b border-b-solid border-cardbordercolor tp:py-1p tp:px-1.5p tp:border-b-0'):
        
        for rate in house_price.find_all('div',class_='flex flex-col w-33pe items-center bo tp:w-half po:w-full border-r border-r-solid border-card-overview-border-color last:border-r-1'):
            price.append((rate.find('span')).text)
            
        for div in house_price.find_all('div',class_='flex flex-col w-33pe items-center bo tp:w-half po:w-full border-r border-r-solid border-card-overview-border-color last:border-r-1'):
            sibling = div.find_next_sibling('div')
            emi.append((sibling.find('div',class_='font-semi-bold heading-6')).text)
        
    
    house_title = [name.split("For Sale  In")[0].strip() for name in house_names]
    area = [name.split("For Sale  In")[-1].strip() for name in house_names]
     
    price = [amt for amt in price if amt != '₹']
    emi = [amt for amt in emi if 'sqft' not in amt]
    
  
    df=pd.DataFrame({'Title':house_title,'Area':area,'Location': location, 'Price':price, 'EMI Details':emi})
    return df
                       
          
    
await NoBroker('https://www.nobroker.in/property/sale/bangalore/multiple?searchParam=W3sibGF0IjoxMi45NzgzNjkyLCJsb24iOjc3LjY0MDgzNTYsInBsYWNlSWQiOiJDaElKa1FOM0dLUVdyanNSTmhCUUpyaEdEN1UiLCJwbGFjZU5hbWUiOiJJbmRpcmFuYWdhciJ9LHsibGF0IjoxMi45MzA3NzM1LCJsb24iOjc3LjU4MzgzMDIsInBsYWNlSWQiOiJDaElKMmRkbFo1Z1ZyanNSaDFCT0FhZi1vcnMiLCJwbGFjZU5hbWUiOiJKYXlhbmFnYXIifSx7ImxhdCI6MTIuOTk4MTczMiwibG9uIjo3Ny41NTMwNDQ1OTk5OTk5OSwicGxhY2VJZCI6IkNoSUp4Zlc0RFBNOXJqc1JLc05URy01cF9RUSIsInBsYWNlTmFtZSI6IlJhamFqaW5hZ2FyIn1d&radius=2.0&city=bangalore&locality=Indiranagar,Jayanagar,Rajajinagar')
```

Question 6: Write a python program to scrape first 10 product details which include product name, price, Image URL from https://www.bewakoof.com/bestseller?sort=popular

```python
async def Bewakoof_product(url):
    
    session = AsyncHTMLSession()
    response = await session.get(url)
    await response.html.arender(sleep=10)
    
    soup = BeautifulSoup(response.html.html,'lxml')
    
    product_name=[]
    product_price=[]
    img_url=[]
    
    
    for name in soup.find_all('div',class_='productNaming bkf-ellipsis'):
        product_name.append((name.find('h2')).text)
    for price in soup.find_all('div',class_='productPriceBox d-flex align-items-end false'):
        product_price.append((price.find('div',class_='discountedPriceText clr-p-black false')).text)
    for image in soup.find_all('img'):
        img_url.append(image['src'])
        
    
    product_name=product_name[:10]
    product_price=product_price[:10]
    img_url=img_url[:10]
       
        
    df=pd.DataFrame({'Product Name':product_name,'Price':product_price,'Image URL':img_url})
    return df
    
    
    
await Bewakoof_product('https://www.bewakoof.com/bestseller?sort=popular%20')
```
