```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')
```

1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
def Most_Viewed_YouTube_Wiki(url):
    
    driver=webdriver.Chrome()
    driver.get(url)
    
    wait=WebDriverWait(driver,10)
    
    #Declaring all the required variables list
    rank = []
    video_name = []
    artist = []
    views = []
    date = []
    
    #Code to locate all the rows of the required data to scrap
    table_rows=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr')))
    
    #Iteraring through all the rows of the 'table_rows' element
    for index,row in enumerate(table_rows):
        
        #Code to locate the column elements
        column=row.find_elements(By.XPATH,'.//td')
        
        #Appending rank in the rank list
        rank.append(index+1)
        
        #Scraping video name 
        video_name.append(column[0].text)
        
        #Scraping artist name 
        artist.append(column[1].text)
        
        #Scraping views name 
        views.append(column[2].text)
        
        #Scraping date name 
        date.append(column[3].text)
        
        
        video_name=[item.split('[')[0] for item in video_name]
        
    #Creating a DataFrame for all the scraped data
    df_most_viewed=pd.DataFrame({'Rank':rank,'Video Name':video_name,'Artist':artist,'Views (Billions)':views,'Date':date},index=range(1,len(rank)+1))
    
    #return the DataFrame
    return df_most_viewed
    
    
Most_Viewed_YouTube_Wiki('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Series
B) Place
C) Date
D) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
def BCCI_International_Fixtures(url):
    
    driver=webdriver.Chrome()
    driver.get(url)
    
    #Declaring all the required variables list
    series_names = []
    places_of_match = []
    date_of_match = []
    time_of_match = []
    
    wait=WebDriverWait(driver,10)
    
    #Code to locate the 'Fixtures' link tab and click on it
    fixtures_link=wait.until(EC.visibility_of_element_located((By.XPATH,'//div[@id="imw-international-men"]//a[@data-event_context="header" and contains(text(),"Fixtures")]')))
    fixtures_link.click()
    time.sleep(5)
    
    #Code to locate all the match_card element which contains all the required data and assigning it to a variable
    match_card_frames=driver.find_elements(By.XPATH,'//div[@class="match-card match-card-fw match-card-up ng-scope"]')
    
    time.sleep(5)
    #Iterating through each match_card_frames elements 
    for each_match_card in match_card_frames:
        
        #Code to locate the series names element and append it to 'series_names' variable
        series=each_match_card.find_element(By.XPATH,'.//h5[@class="match-tournament-name ng-binding"]')
        series_names.append(series.text)
        
        #Code to locate the date of match element and append it to 'date_of_match' variable
        date=each_match_card.find_element(By.XPATH,'.//div[@class="match-dates ng-binding"]')
        date_of_match.append(date.text)
        
        
        #Code to locate the time of match element and append it to 'time_of_match' variable
        times=each_match_card.find_element(By.XPATH,'.//div[@class="match-time no-margin ng-binding"]')
        time_of_match.append(times.text)
        
        #Code to locate the place of match element and append it to 'place' variable
        all_location=each_match_card.find_elements(By.XPATH,'.//div[@class="match-place ng-scope"]/span')
        location=[loc.text.strip() for loc in all_location]
        complete_location=' '.join(location)
        
        places_of_match.append(complete_location)

        
    
    #Creating a DataFrame for all the scraped data
    df_bcci_match=pd.DataFrame({'Series':series_names,'Location':places_of_match,'Date':date_of_match,'Time':time_of_match},index=range(1,len(series_names)+1))
    
    #return the DataFrame
    return df_bcci_match
    
    
BCCI_International_Fixtures('https://www.bcci.tv/')
```

3. Scrape the details of State-wise GDP of India from statisticstime.com. url = http://statisticstimes.com/
You have to find following details: 

A) Rank

B) State

C) GSDP(23-24)- at current prices

D) GSDP(22-23)- at current prices

E) Share(21-22)

F) GDP($ billion)

```python
def State_wise_GDP_Of_India(url):
    
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    
    driver=webdriver.Chrome(options=chrome_options)
    driver.get(url)
    
    #Declaring all the required variables list
    rank_of_states = []
    state_name = []
    gsdp_24_23 = []
    gsdp__23_22 = []
    states_shares = []
    gdp_of_states = []
   
    wait=WebDriverWait(driver,10)
    
    #Code to locate the economy tab link and move the mouse cursor to it
    economy_button_link=wait.until(EC.visibility_of_element_located((By.XPATH,'//button[@class="dropbtn" and contains(text(),"Economy")]')))
    economy_button_link.click()
    
    #Code to find the 'India' link from 'Economy' tab and move the coursor to it and click on it
    india_link = wait.until(EC.visibility_of_element_located((By.XPATH,'//div[@class="dropdown-content"]/a[@href="economy/india-statistics.php"]')))
    india_link.click()
    
    #Code to located the 'GDP of Indian states' link and click on it
    gdp_of_indian_states_link=wait.until(EC.visibility_of_element_located((By.XPATH,'//a[@href="india/indian-states-gdp.php" and contains(text(),"GDP of Indian states")]')))
    gdp_of_indian_states_link.click()
    
    
    #Code to locate all the rows of required data in the table
    gdp_table_body_row=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//table[@id="table_id"]/tbody/tr')))
    
    #Iterating through each row in a table
    for rows in gdp_table_body_row:
        
        #Locating each column in rows
        column=rows.find_elements(By.XPATH,'.//td')
        
        rank_of_states.append(column[0].text)
        state_name.append(column[1].text)
        gsdp_24_23.append(column[2].text)
        gsdp__23_22.append(column[3].text)
        states_shares.append(column[5].text)
        gdp_of_states.append(column[6].text)
        
    
    #Creating a DataFrame
    df_indian_states_gdp=pd.DataFrame({'Rank':rank_of_states,'State':state_name,'GSDP (24-23)':gsdp_24_23,'GSDP (23-22)':gsdp__23_22,'State\'s share':states_shares,'GDP':gdp_of_states},index=range(1,len(rank_of_states)+1))
    
    #Returning DataFrame
    return df_indian_states_gdp
    
    
    
State_wise_GDP_Of_India('https://www.statisticstimes.com/')
```

4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
def Github_Trending_Repositories(url):
    
    
    driver=webdriver.Chrome()
    driver.get(url)
    
    #Declaring all the required variables list
    repository_title = []
    repository_description = []
    repository_urls = []
    contributors_count = []
    language_used = []
   
    
    wait=WebDriverWait(driver,10)
    
    #Locate 'Open Source' element tab link and click on it
    open_source_tab=wait.until(EC.visibility_of_element_located((By.XPATH, '//button[@type="button" and contains(text(),"Open Source")]')))
    open_source_tab.click()
    
    #Locate 'Trending' element link and click on it
    trending_repository_link=wait.until(EC.visibility_of_element_located((By.XPATH, '//a[contains(text(),"Trending")]')))
    trending_repository_link.click()
    
    #Code to find elements where all the elements are present and assigning it to a variable 'trending_repository_frames'
    trending_repository_frames=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//article[@class="Box-row"]')))
    
    #Iterating through each trending_repository_frames
    for each_frame in trending_repository_frames:
        
        #Locate the title element and append it to repository_title variable
        title=each_frame.find_element(By.XPATH,'.//h2[@class="h3 lh-condensed"]/a')
        repository_title.append(title.text.strip())
        
        #Locate the description element and append it to repository_description variable under try...except block
        try:
            description=each_frame.find_element(By.XPATH,'.//p[@class="col-9 color-fg-muted my-1 pr-4"]')
            repository_description.append(description.text)
        except NoSuchElementException:
            repository_description.append('-')
        
        #Locate the url element and append it to repository_urls variable
        url_element=each_frame.find_element(By.XPATH,'.//h2[@class="h3 lh-condensed"]/a')
        repository_urls.append(url_element.get_attribute('href'))
    
    #Iterate through all the repository_urls
    for urls in repository_urls:
        
        #Load each url in the browsers
        driver.get(urls)
        
        #Locate the contributors count element and append it to contributors_count variable under try...except block
        try:
            contributors = driver.find_element(By.XPATH,'//h2[@class="h4 mb-3"]/a[contains(text(),"Contributors")]/span[@class="Counter ml-1"]')
            contributors_count.append(contributors.text)
        except NoSuchElementException:
            contributors_count.append('-')
        
        #Locate the language used element and append it to language_used variable under try...except block
        try:
            languages=driver.find_elements(By.XPATH,'//div[@class="BorderGrid-cell"]/ul/li[@class="d-inline"]/a/span[@class="color-fg-default text-bold mr-1"]')
            language=[lang.text.strip() for lang in languages]
            lang=', '.join(language)
            language_used.append(lang)
        except NoSuchElementException:
            language_used.append('-')
       
    
    #Creating a DataFrame
    df_Github_Trending_Repositories=pd.DataFrame({'Title':repository_title,'Description':repository_description,'No. of contributors':contributors_count,'Language':language_used},index=range(1,len(repository_title)+1))
    
    #Returning a DataFrame
    return df_Github_Trending_Repositories

    
Github_Trending_Repositories('https://github.com/')
```

5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ You have to find the following details:

A) Song name

B) Artist name

C) Last week rank

D) Peak rank

E) Weeks on board

```python
def Billboard_top_100(url):
    
    driver=webdriver.Chrome()
    driver.get(url)
    
    #Declaring all the required variables list
    song_name = []
    artist_name = []
    last_week_rank = []
    peak_rank = []
    weeks_on_board = []
   
    
    wait=WebDriverWait(driver,10)
    
    #Locate 'Billboard Hot 100' tab link and click on it
    billboard_hot_100=wait.until(EC.visibility_of_element_located((By.XPATH,'//li[@class="o-nav__list-item "]/a[contains(text(),"Billboard Hot 100")]')))
    billboard_hot_100.click()
    
    #Code will locate all the table rows
    table_rows=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@class="o-chart-results-list-row-container"]')))
    
    #Iterating through each row 
    for rows in table_rows:
        
        #Code to locate the song name element and append it to variable song_name
        song=rows.find_element(By.XPATH,'.//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li/h3')
        song_name.append(song.text)
        
        #Code to locate the artist name element and append it to variable artist_name
        name=rows.find_element(By.XPATH,'.//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li/h3/following-sibling::span')
        artist_name.append(name.text)
        
        #Code to locate the 'Last week rank' element and append it to variable last_week_rank
        l_w_rank=rows.find_element(By.XPATH,'.//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li[4]/span')
        last_week_rank.append(l_w_rank.text)
        
        #Code to locate the 'Peek rank' element and append it to variable peak_rank
        p_rank=rows.find_element(By.XPATH,'.//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li[5]/span')
        peak_rank.append(p_rank.text)
        
        #Code to locate the 'Weeks on board' element and append it to variable weeks_on_board
        w_on_board=rows.find_element(By.XPATH,'.//ul[@class="lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max"]/li[6]/span')
        weeks_on_board.append(w_on_board.text)
        
        
    #Creating DataFrame    
    df_billboard_top_100=pd.DataFrame({'Songs':song_name,'Artist\'s Name':artist_name,'Last week':last_week_rank,'Peak position':peak_rank,'Weeks on chart':weeks_on_board},index=range(1,len(song_name)+1))
    
    #Return the DataFrame
    return df_billboard_top_100    
        
        
Billboard_top_100('https://www.billboard.com/')
```

6. Scrape the details of Highest selling novels.

A) Book name

B) Author name

C) Volumes sold

D) Publisher

E) Genre

Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare

```python
def Highest_Selling_Novels(url):
    
    
    driver=webdriver.Chrome()
    driver.get(url)
    
    #Declaring all the required variables list
    rank = []
    book_name = []
    author_name = []
    volumes_sold = []
    publisher = []
    genre = []
   
    
    wait=WebDriverWait(driver,10)
    
    #Code to locate the table rows
    table_rows=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//table[@class="in-article sortable"]/tbody/tr')))
    
    #Iterating through each rows
    for rows in table_rows:
        
        #Locate all the column elements
        columns=rows.find_elements(By.XPATH,'.//td')
        
        rank.append(columns[0].text)
        book_name.append(columns[1].text)
        author_name.append(columns[2].text)
        volumes_sold.append(columns[3].text)
        publisher.append(columns[4].text)
        genre.append(columns[5].text)
        
    #Creating a DataFrame
    df_highest_selling_novels=pd.DataFrame({'Rank':rank,'Title':book_name,'Author':author_name,'Volume sales':volumes_sold,'Publisher':publisher,'Genre':genre},index=range(1,len(rank)+1))
    
    #Returning a DataFrame
    return df_highest_selling_novels
    
Highest_Selling_Novels('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

8. Details of Datasets from UCI machine learning repositories.

Url = https://archive.ics.uci.edu/ You have to find the following details:
        
A) Dataset name

B) Data type

C) Task

D) Attribute type

E) No of instances

F) No of attribute 

G) Year

```python
def UCI_Machine_Learning_Repositories(url):
    
    driver=webdriver.Chrome()
    driver.get(url)
    
    #Declaring all the required variables list
    dataset_name = []
    data_type = []
    associated_task = []
    attribute_type = []
    no_of_instances = []
    no_of_attribute = []
    donated_year = []
    url_links = []
    
    wait=WebDriverWait(driver,10)
    
    #Locate all the url elements and assigning it to a variable urls_elements
    urls_elements=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    
    #Iterating through each URL elements
    for urls in urls_elements:
        
        #Extract the href from url elements and append it to url_links variable
        url_links.append(urls.get_attribute("href"))
        
    
    #Iterate through each url link
    for url in url_links:
        
        #Load each url in the browser
        driver.get(url)
        
        #Locate the dataset name element and append it to dataset_name variable under try...except block
        try:
            name=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
            dataset_name.append(name.text)
        except NoSuchElementException:
            dataset_name.append('-')
        
        #Locate the data type element and append it to data_type variable under try...except block
        try:
            d_type=driver.find_element(By.XPATH,'//h1[@class="text-lg font-semibold text-primary" and contains(text(),"Dataset Characteristics")]/following-sibling::p')
            data_type.append(d_type.text)
        except NoSuchElementException:
            data_type.append('-')
        
        #Locate the associated task element and append it to associated_task variable under try...except block
        try:
            task=driver.find_element(By.XPATH,'//h1[@class="text-lg font-semibold text-primary" and contains(text(),"Associated Tasks")]/following-sibling::p')
            associated_task.append(task.text)
        except NoSuchElementException:
            associated_task.append('-')
        
        #Locate the attribute type element and append it to attribute_type variable under try...except block
        try:
            a_type=driver.find_element(By.XPATH,'//h1[@class="text-lg font-semibold text-primary" and contains(text(),"Feature Type")]/following-sibling::p')
            attribute_type.append(a_type.text)
        except NoSuchElementException:
            attribute_type.append('-')
        
        #Locate the instances element and append it to no_of_instances variable under try...except block
        try:
            n_of_instance=driver.find_element(By.XPATH,'//h1[@class="text-lg font-semibold text-primary" and contains(text(),"Instances")]/following-sibling::p')
            no_of_instances.append(n_of_instance.text)
        except NoSuchElementException:    
            no_of_instances.append('-')
        
        #Locate the Features element and append it to no_of_attribute variable under try...except block
        try:
            n_of_attribute=driver.find_element(By.XPATH,'//h1[@class="text-lg font-semibold text-primary" and contains(text(),"Features")]/following-sibling::p')
            no_of_attribute.append(n_of_attribute.text)
        except NoSuchElementException:
            no_of_attribute.append('-')
        
        #Locate the donated year element and append it to donated_year variable under try...except block
        try:    
            year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
            donated_year.append(year.text)
        except NoSuchElementException:
            donated_year.append('-')
    
    #Creating a DataFrame and assigning it to a variable df_ml_repo
    df_ml_repo=pd.DataFrame({'Dataset Name':dataset_name,
                            'Data Type':data_type,
                            'Associated Tasks':associated_task,
                            'Feature Type':attribute_type,
                            'Instance':no_of_instances,
                            'Features':no_of_attribute,
                            'Donated Year':donated_year},index=range(1,len(dataset_name)+1))
    
    #Return DataFrame
    return df_ml_repo



UCI_Machine_Learning_Repositories('https://archive.ics.uci.edu/')
```
